```python
class ValidationError(Exception):
    """Exception raised for validation errors."""
    def __init__(self, feild, message="Invalid value"):
        self.feild = feild
        self.message = message

try:
    raise ValidationError("username", "Must not be empty")
except ValidationError as e:
    # print("Error handled")
    print(e)
```

```python
class ValidationError(Exception):
    """Exception raised for validation errors."""
    def __init__(self, feild, message="Invalid value"):
        self.feild = feild
        self.message = message

try:
    raise ValidationError("username", "Must not be empty")
except ValidationError as e:
    print("Error handled")
    print(e)
```

```python
class ApplicationError(Exception):
    """Base class for all application exceptions."""
    pass
class NetwrokError(ApplicationError):
    """Exception raised for network-related errors."""
    pass
class FileError(ApplicationError):
    """Exception raised for file-related errors."""
    pass
try:
    raise NetwrokError("Network is unreachable")
except ApplicationError as e:
    print(f"Caught an application error: {e}")
```

# 2.Writing to a CSV File

```python
data = [
    ['Name', 'Age', 'City'],
    ['Alice', 30, 'New York'],
    ['Bob', 25, 'Los Angeles'],
    ['Charlie', 35, 'Chicago']
]
with open('output.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```

```python
data = [
    ['Name', 'Age', 'City'],
    ['Alice', 30, 'New York'],
    ['Bob', 25, 'Los Angeles'],
    ['Charlie', 35, 'Chicago']
]
with open('output.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
    for row in data:
        writer.writerow(row)
```

```python
data = [
    {'Name': 'Alice', 'Age': 30, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 25, 'City': 'Los Angeles'},
    {'Name': 'Charlie', 'Age': 35, 'City': 'Chicago'}
]

with open('output.csv', mode='w', newline='') as file:
    fieldnames = ['Name', 'Age', 'City']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)
```

```python
fieldnames = ['Name', 'Age', 'City']
writer = csv.DictWriter(open('output.csv', mode='w', newline=''), fieldnames=fieldnames)
writer.writeheader()
writer.writerows(data)  
```
